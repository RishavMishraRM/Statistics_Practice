The weight of 25 obese people were taken before enrolling them into the nutrition camp. The population mean weight is found to be 45kg before starting the camp. 
After finishing the camp, for the same 25 people, the sample mean was found to be 75 with a standard deviatoon of 25. Did the fitness camp work ?

H0 ===> mu=45(sample_mean is a true/population mean)

H1 ===> mu != 45(two tail test) (sample_mean is not a true/population mean)

```python
n = 25
population_mean = 45
sample_mean = 75 
sample_std = 25
```

```python
t_stats = (sample_mean - population_mean)/(sample_std/np.sqrt(n))
```

```python
t_stats
```

```python
df = n-1
```

```python
df
```

```python
alpha = 0.05
```

```python
CI = 0.95
```

```python
critical_value = stats.t.ppf(CI, df)
```

```python
critical_value
```

```python
print("t-stats : ", t_stats)
print("critical value : ", critical_value)
```
